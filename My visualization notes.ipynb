# Lineplot
used for x and y continuous variables , we can use multiple lines in one graphs of similar measure unite <br>
Line graphs provide an excellent way to map independent and dependent variables that are both quantitative <br>
Most appropriate for the time series data

```python
# using subplot for multiple plots

# subplot(rows , columns, figsize , sharey , sharex, hspace = 0.1 )
fig, axes = plt.subplots(2, 3, figsize=(15, 10), sharey=True, sharex = True )
fig.suptitle('Lineplot with different options')
fig.tight_layout() # spacing between the graphs

# using x = year and y = kilometer driven

sns.lineplot(ax = axes[0,0], data = df, x = 'Year', y = 'Kilometers_Driven',style = 'Fuel_Type', hue = 'Fuel_Type')
sns.lineplot(ax = axes[0,1], data = df, x = 'Year', y = 'Kilometers_Driven',size = 'Fuel_Type')
sns.lineplot(ax = axes[0,2], data = df, x = 'Year', y = 'Kilometers_Driven', marker = True, dashes=False, hue = 'Fuel_Type')
sns.lineplot(ax = axes[1,0], data = df, x = 'Year', y = 'Kilometers_Driven', markers = True, dashes = False, style = 'Fuel_Type')
sns.lineplot(ax = axes[1,1], data = df, x = 'Year', y = 'Kilometers_Driven', err_style="bars", ci=95)

# Parameters:
# ax = axes[rw,col] - grpha position in subplot            
# data = dataframe
# x = continueos variable on x axis
# y = contineous variable on y axis
# Assign a grouping semantic (hue, size, or style) to plot separate lines two or more sematics can also be used for one column
# hue = categorical column to draw lines based on color
# size = categorical column to draw lines based on line width
# style = categoricla column to draw lines based on style
# markers = True/False - make market visible (can only be used with style sematic)
# dashes = True/False - remove the dashesh (can only be used with style sematic)
# Show error bars instead of error bands and plot the 95% confidence interval (standard error)
# err_style = style of graphic showing error ( band or bars) default is band
# ci = confidence interval in percent 

plt.show()
```

# Barplot

It shows the relationship between a numeric and a categoric variable.<br>
Mostly contineour variable is used on the y axis of the barplot and catergorical on X axis

categorical columns : 'Name', 'Location','Fuel_Type', 'Transmission', 'Owner_Type', 'Seats' <br>
nuerical columns : 'Kilometers_Driven' , 'Year',  'Mileage', 'Engine', 'Power', 'Price'

```python
# creating the subplots to show figures

fig, axes = plt.subplots(2,3, figsize = (20,15), sharex = False, sharey = False)
fig.suptitle('Barplot with different options')

sns.barplot(ax = axes[0,0], data = df, x = 'Location', y = 'Price' , palette = 'cubehelix')
sns.barplot(ax = axes[0,1], data = df, x = 'Fuel_Type', y = 'Mileage', hue = 'Transmission')
sns.barplot(ax = axes[0,2], data = df, x = 'Fuel_Type', y = 'Mileage', order = ['Petrol','Diesel','LPG'])
sns.barplot(ax = axes[1,0], data = df, x = 'Fuel_Type', y = 'Mileage', estimator = np.median)
sns.barplot(ax = axes[1,1], data = df, x = 'Transmission', y = 'Mileage', palette='coolwarm', ci = None)
sns.barplot(ax = axes[1,2], data = df, x = 'Price', y = 'Location' , palette = 'cubehelix', orient='h')

# Parameters - 
# ax = axes[rw,col] - grpha position in subplot            
# data = dataframe
# x = Categorical variable on x axis
# y = contineous variable on y axis
# hue = categorical columns to differentiate
# order = order of categorical column on x axis [a list] , you can also limit the variables
# estimator = median, mean ( central tendency estimator) need to use numpy function mean/median
# palette= color scheme of bars. eg. icefire , vlag , coolwarm
# palette options > Blues_d , rocket, cubehelix , mako , flare , crest , magma , viridis , rocket_r ,seagreen
# orient='h' - changes the orientation of bar plot to horizontal, x axis need to be continuous variable
# ci = standard error (at confidence interval, 0-99% , None to hide the error bar)
# In seaborn its hard to make stacked barplot with hue - you can use plt.plot(kind = 'bar', staked = True ,other para)

plt.show()
```

# Histogram - histplot

The histogram is used for variables whose values are numerical and measured on an interval scale. Used to check distribution and outliers <br>


categorical columns : 'Name', 'Location','Fuel_Type', 'Transmission', 'Owner_Type', 'Seats' <br>
nuerical columns : 'Kilometers_Driven' , 'Year',  'Mileage', 'Engine', 'Power', 'Price'

```python
# creating the subplots to show figures

fig, axes = plt.subplots(3,3, figsize = (20,15), sharex = False, sharey = False)
fig.suptitle('Histplot with different options')

# ploting continuous variables opposite to count

sns.histplot(ax = axes[0,0] , data = df, x = 'Mileage', bins= 20, stat = 'frequency')
sns.histplot(ax = axes[0,1] , data = df, x = 'Price' , binwidth=2, binrange= (0,50))
sns.histplot(ax = axes[0,2] , data = df, x = 'Power', kde = True, bins = 30 ,  hue = 'Owner_Type')
sns.histplot(ax = axes[1,0] , data = df, x = 'Mileage',hue = 'Owner_Type', binwidth=2, multiple= 'stack')
sns.histplot(ax = axes[1,1] , data = df, x = 'Engine',hue = 'Fuel_Type', bins= 20, binrange = (500,4000) , element= 'step')
sns.histplot(ax = axes[1,2] , data = df, x = 'Power',hue = 'Transmission', element= 'poly')
sns.histplot(ax = axes[2,0] , data = df, x = 'Mileage', element= 'step', stat = 'density')
sns.histplot(ax = axes[2,1] , data = df, x = 'Mileage', fill = False )
sns.histplot(ax = axes[2,2] , data = df, x = 'Mileage', fill = False, hue = 'Transmission', cumulative=True, element='step' )

# Parameters - 
# ax = axes[rw,col] - grpha position in subplot            
# data = dataframe
# x = Categorical variable on x axis
# y = contineous variable on y axis
# hue = categorical columns to differentiate
# binwidth - width of the column
# bins = number of bins to show on graph
# binrange = (start, end) range of numbers 
# kde = True/False - kde kernel density estimate line
# multiple = 'stack' , default hue is layered , we can make it stack
# element = make it easy see overlaping hue bars, step - steps, poly - kdeline
# stat= y axis function ( density , probability, frequency , count )
# fill = False - remove the color fill
# cumulative = True - cumulative counts

plt.show()
```

# Catplot

Catplot show the relationship between a numerical and one or more categorical variables using one of several visual representations. <br> Catplot is mostly used to show the multiple plots of same kind for unique elements in the categorical columns

<ul>Categorical Scatterplots:
    <li>stripplot() (with kind="strip"; the default)</li>
    <li>swarmplot() (with kind="swarm")</li>
</ul>
<ul>Categorical distribution plots:
    <li>boxplot() (with kind="box")</li>
    <li>violinplot() (with kind="violin")</li>
    <li>boxenplot() (with kind="boxen")</li>
</ul>
<ul>Categorical estimate plots:
    <li>pointplot() (with kind="point")</li>
    <li>barplot() (with kind="bar")</li>
    <li>countplot() (with kind="count")</li>
</ul>

categorical columns : 'Name', 'Location','Fuel_Type', 'Transmission', 'Owner_Type', 'Seats' <br>
nuerical columns : 'Kilometers_Driven' , 'Year',  'Mileage', 'Engine', 'Power', 'Price'

```python
g = sns.catplot(kind = 'violin',data = df, x = 'Owner_Type', y = 'Mileage', col = 'Fuel_Type', row = 'Transmission', aspect = 0.8, height= 5)

# kind = anykind of graph can be used from the list above (except count it takes only one argument x)
# col = col name to make graph columns based on 
# row = row name to make grpah rows based on  
# col_wrap - maximum number of plot in a row , next will be move to next row
# col_wrap and row cannot be used together. 
# height = height of plots
# aspect = 0-1 , aspect ration of x and y 
# orient = 'h' for horizontal orientation

(g.set_axis_labels("", "")
  .set_xticklabels(["First", "Second", "Third","Forth more"])
  .set_titles("{col_name} {col_var}")) 

# setting the properties of label 
```

# Jointplot

Draw a plot of two numerical variables with bivariate and univatiate graphs with hue as categorical variable

```python
sns.jointplot( x ='Power' , y = 'Mileage', hue = 'Transmission', data = df, kind = 'scatter')
sns.jointplot( x ='Power' , y = 'Mileage', data = df, kind = 'hex', joint_kws = dict(gridsize = 20))
sns.jointplot( x ='Power' , y = 'Mileage', data = df, kind = 'reg', marker="+")
sns.jointplot( x ='Power' , y = 'Mileage', data = df, kind = 'resid',marginal_kws=dict(bins=25, fill=False) )
sns.jointplot( x ='Power' , y = 'Mileage', hue = 'Transmission', data = df, kind = 'kde')


# x and y are axes for elements to be compared
# data = dataset 
# kind = scatter, hex, reg, resid, kde
# color = color of plot
# height = figure height
# ration = aspect ration of plot
# dropna = True/False , do not show nan if True
# joint_kws = dict(gridsize = 20) - in hex plot gridzise is number of hexs on the plot
# marker = marker type
# marginal_kws=dict(bins=25, fill=False) - related to margin resid

plt.show()
```

# Heatmap

Heatmaps are used to show relationships between two variables, one plotted on each axis. <br>
The variable can be of any type

```python
annot_kws={'fontsize':20, 
           'fontstyle':'italic',  
           'color':"k",
           'alpha':0.6, 
           'rotation':"horizontal",
           'verticalalignment':'center',
           'backgroundcolor':'r'}

# Everything above for the annot is self explainatory. 

plt.figure(figsize = (15,15))

sns.heatmap(df.corr(), vmin = -1, vmax = 1, center = 0, annot = True, linewidths = 0.25, cmap = 'YlGnBu', cbar_kws={"orientation": "horizontal"})

# dataframe or pivot table
# vmin / vmax - min max on color scale
# center - center point specified if we want any custom
# annot - denote value in each box
# fmt - format of the interger f for float d for int
# annot_kws = annot_kws - annot variable as dictionary. 
# linewidths - line seperating boxes
# cmap - color scheme colormap (YlGnBu,Blues ,BuPu,  Greens, PiYG )
# yticklabels = True, False or alternate number eg. if 2 passed then show every second label
# xticklabels = same as y ticks
# cbar=False - to hide the color bar
# cbar_kws={"orientation": "horizontal"} - color bar orientation 
 
```
